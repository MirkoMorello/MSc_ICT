# Speech processing

4 Khz/8 Khz are good sampling rate for voice, which is what it's used for voice telephone calls. Music is at 44 Khz. Cd's are formatted like this. Also consumer devices like Alexa has this sample rate.  
- Pitch: Higher frequencies perceived as higher pitch, hear sounds in 10-20.000 Hz range
- Loudness: Higher amplitude results in louder sounds, measured in decibels (db), 0 db represents hearing threshold.
- Timbre: Complex patterns added to the lowest, or fundamental, frequency of a sound, referred to as spectra, spectra enables us to distinguish musical instruments, characteristics of sound that allows us to differentiate between sound with the same putch and loudness but come from different sources.  


Intensity (i) of a wave is the rate at which sound energy flows through a unit area (A) perpendicular to the direction of travel.  
Decibels describe the intensity relative to thereshold of hearing based on multiples of 10.  

Perception of human it's not of the same decibels a different frequencies, it's not linear.

### Speech recognition:
refers to converting an acoustic waveform of spoken speech to the corresponding text (speech-to-text).

### Speaker recognition and speaker verification:
refer to, respectively, identifying the speaker (who is speaking?) and  verifying whether the speaker is who he claims to be (is it really you?).

### Speech synthesis:
creation of a natural sounding speech signal from text input (text-to-speech).

### Speech enhancement:
refers to improving a recorded speech signal, for example with the objective of removing background noise (noise attenuation) or the effect of room acoustics.

### Wake-word and keyword detection
Refers to the task where the purpose is to find single characterizing words from continuous speech

### Voice Activity Detection (VAD)
Refers to the task of determining whether a signal contains speech or not (is someone speaking?).

### Speech diarisation
is the process of segmenting a multi-speaker conversation into continuous single-speaker segments.

# Voice Activity Detection (VAD)

We calculate an average of variation of sound, this is because we want to avoid to capture noise that usually has a lower air life span.  
We have a threshold that let's us know if we've just detected speech.  



we need advanced VAD methods for noisy speech, we need to identify the characteristics that differentiate between speech and noise, measures for such characteristics are known as features.  
Signal to noise ratio:
$$\text{SNR} = \frac{P_{\text{signal}}}{P_{\text{noise}}}$$

#### MFCC
Audio extraction geatures in audio, can be used as a description of envelope information and it is thus a useful set of features.  
It's a kind of spectrogram, the only differences is in the scale and in the use of the last module which uses Discrete Fourier Transforms (DFT)  
We use a DFT, then we change the scale and end up in the mel scale (+ log), which shows us better understanding of human perception of speech, then we go back again to time domain with a DFT which uses cosine (DCT), here we are able to study the sound and understand where are these specific sounds (human voices) that we wanted to find all along.  

We end up with either an image with MFCC that we can use, or we do derivatives and other operations and end up with coefficients, it depends on the kind of application we are applying.  
Usually, if we are working with ML we just use the image, since we don't mind a complex information, while for devices like alexa etc that have narrowed capabilities, we process the information with derivatives etc in order to do VAD.
