# Speaker Recognition

It subdivides into 3 different tasks that require 3 different solutions:
- Speaker identification, it has to understand the voice of a specific person out of potentially multiple voices, at the end of the day it's a classification task.
- Speaker Detection/tracking.
- Speaker Verification, is the task of accepting or rejecting the identity claim of a speaker by analysing their acoustic samples. This isn't exactly a classification task, it's a bit different, we model the model from the entirity of possible malintentioned people, and then on the person that has to be authorized, and we develop a strategy from there.  

## Speaker recognition and verification
It needs an open set, so I can add people and have a rejection mechanism, to do so, we basically calculate the entropy of the logits in the last layer and put a threshold on it.  
How to calculate the threshold? in the roc curve, in the x axis, we draw an horizontal line from the False Positive Rate at 3% and find the threshold on the True Positive Rate.  

The second strategy, instead of thresholding, we basically model the rest of the world, I train the model in N+1 model, the 1 is the rest of the world, this implies we need samples from other people.

#### Verification decision apporaches have roots in signal detection theory
- 2 class hypothesis test:
  - H0: the spearker is an impostor
  - H1: the speker is indeed the claimed speaker
- Statistic computed on test utterance S as likelihood ratio:

$$A = \log{\frac{\text{likelihood S came from speaker model}}{\text{Likelihood S did not come from speakerr model}}}$$

Verification: usual apporach is via likelihood ratio tests, hypothesis testing, i.e. By Bayes:
$$\frac{P(\text{target}|x)}{P(\text{nontarget}|x)} = \frac{P(x|\text{target}) P(\text{target})}{P(x|\text{nontarget})P(\text{nontarget})}$$
