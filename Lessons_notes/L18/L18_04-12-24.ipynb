### Classifier based Personalization
I create many sub models one for each cluster of users.
- Single personalization strategy: a new users is matched with one of the users in the training, meaning I found his cluster and his personalized model
- Multi-personalization strategy: the new user is matched with multiple clusters, so multiple models, and the classification comes from an average of the classifiers output.

Steps are the following:
- General purpose classifier
- Calculate embedding, and use this embedding to find similarities
- Fine tune the general purpose classifier with this new data calculated with similarieties
- Use this new model to classify actions during inference

Hybrid means retraining the classifier, if we use also the similarity it improves of 11%
